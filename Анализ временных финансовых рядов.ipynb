## Предобработка данных

```python
df = pd.DataFrame()
df['TRADEDATE'] = raw_df[['TRADEDATE']]
```

```python
df.drop_duplicates(keep='first', inplace=True)
```

```python
df_p = raw_df.loc[raw_df.SECID == 'GAZP', ['CLOSE', 'TRADEDATE']]
```

```python
result.rename(columns = {'CLOSE' : 'GAZP'}, inplace = True)
```

```python
for ticker in tickers[1:]:
    df_p = raw_df.loc[raw_df.SECID == ticker, ['CLOSE', 'TRADEDATE']]
    result = pd.merge(result, df_p, on='TRADEDATE', how='left')
    result.rename(columns = {'CLOSE' : ticker}, inplace = True)
result.head()
df = result
```

```python
df.head()
```

NaN значения появились в датафрейме в связи с отсутствием данных о торгах акций некоторых компаний в определенный день. Для начала, найдем процент NaN значений по столбцам

```python
nans = dict(round(df.isna().mean()*100))
tickers_to_del = []
for key in nans.keys():
    if (nans[key]>=80):
        tickers_to_del.append(key)
nans
tickers_to_del
```

Как видно, у тикеров 'GLTR', 'OZON', 'HHRU','FIXP' количество NaN ячеек превышает 80%. Удаляем соответствующие стобцы из датафрейма, чтобы очистить данные

```python
for ticker in tickers_to_del:
    df.pop(ticker)
```

Найдем столбцы с Nan значениями

```python
nans = dict(round(df.isna().mean()*100, 3))
tickers_to_fill = []
for key in nans.keys():
    if (nans[key]>0):
        tickers_to_fill.append(key)
tickers_to_fill
```

Заполняем NaN значения в столбцах TCSG, FIVE, MAIL, RUAL, DSKY, POGR на среднее по соответствующему столбцу и проверяем, удалось ли избавиться NaN

```python
for ticker in tickers_to_fill:
    mean_value = df[ticker].mean()
    df[ticker] = df[ticker].fillna(mean_value)
df.isna().sum()
```

Теперь перейдем к созданию дополнительных столбцов. В первую очередь, переводим время в тип datetime64[ns] для удобной работы со столбцом

```python
df['TRADEDATE'] = pd.to_datetime(df.TRADEDATE) 
```

Добавляем столбец дня недели

```python
df['DAY_NAME'] = df.TRADEDATE.dt.day_name()
```

Проверяем датафрейм на наличие выходных дней. Необходимо создать столбец со значениями типа bool

```python
weekday1 = pd.to_datetime('31/12/2021')
weekday2 = pd.to_datetime('1/1/2021')
weekday3 = pd.to_datetime('2/10/2021')
df.loc[df.TRADEDATE == weekday3, :]
df.loc[df['DAY_NAME'] == 'Saturday']
```

В датафрейме оказалось 5 строк с выходным днем. Создаем столбец "WEEKEND"

```python
s = df['DAY_NAME']
is_weekend = []
for k in s:
    if k == 'Saturday':
        is_weekend.append(True)
    else:
        is_weekend.append(False) 
print(len(is_weekend))
df['WEEKEND'] = is_weekend     
```

Создаем столбец "MONTH_END". Концом месяца считаются 25-31 числа

```python
s = df.TRADEDATE.dt.day
is_month_end = []
for i in s:
    is_month_end.append(i in range(25,32))
print(len(is_month_end))
df['MONTH_END'] = is_month_end
```

Создаем столбец "MONTH_NAME"

```python
month_name = ('January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October',
              'November', 'December')
month = []
for i in df.TRADEDATE.dt.month:
    i -= 1
    month.append(month_name[i])
df['MONTH_NAME'] = month
```

### Группировка датафрейма по месяцам для исследования средних изменений цен акций в рассматриваемые периоды

```python
df_month_change = df.groupby('MONTH_NAME').agg('mean')
df_month_change.pop('MONTH_END')
df_month_change = df_month_change.T
df_month_change = df_month_change[['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October',
                                   'November', 'December']]
df_month_change = df_month_change.mean().pct_change()
df_month_change.plot(kind='bar')
plt.show()
```

Исходя из диаграммы, можно сделать вывод, что наибольшее падение цен акций наблюдается в начале года. При этом наибольший рост акций наблюдается с октября по декабрь

### Проверка зависимости акций Газпрома, Сбербанка и Яндекса

```python
df_gsy = df[['GAZP', 'SBER', 'SBERP', 'YNDX']]
```

```python
df_gsy = df_gsy.pct_change()*100
df_gsy[['TRADEDATE', 'MONTH_NAME', 'DAY_NAME', 'WEEKEND', 'MONTH_END']] = df[['TRADEDATE', 'MONTH_NAME', 'DAY_NAME', 'WEEKEND', 
                                                                              'MONTH_END']]  
df_gsy['TRADEDATE'] = pd.to_datetime(df_gsy.TRADEDATE) 
years = []
for i in df_gsy.TRADEDATE.dt.year:
    years.append(i)
df_gsy['YEAR'] = years
df_gsy.head()
```

```python
sns.pairplot(df[['GAZP', 'SBER', 'SBERP', 'YNDX']])
plt.show()
```

Исходя из графиков, можно сделать следующие выводы:
- Цена на акции Сбербанка имеет распределение близкое к нормальному
- Акции Газпрома и Сбербанка имеют сильную прямую корреляцию
- Акции Яндекса и Газпрома также имеют корреляцию, но она выражена в меньшей степени

Можно сделать вывод, что в конце месяца наблюдается наибольшая зависимость цен на акции данных компаний

```python
sns.relplot(data=df_gsy, x='YEAR', y='GAZP', kind='line')
sns.relplot(data=df_gsy, x='YEAR', y='SBER', kind='line')
sns.relplot(data=df_gsy, x='YEAR', y='YNDX', kind='line')
plt.show()
```

По графикам можно оценить то, как менялись цены акций за последние 5 лет. Так, в 2017 году поднялись цены акций Яндекса, в то время, как цены Сбербанка и Газпрома упали. При этом в 2018 году падение цен акций всех трех компаний связано в введением антироссийских санкций. В 2019 году компаниям удалось побороть кризис и цены начали расти вплоть до пандемии COVID-19. В пандемию в плюс вышла IT-компания Яндекс, не сильно упал Сбер, но пострадал от кризиса Газпром. В 2021 году цена Сбера практически не менялась, поднялась цена Газпрома в связи с глобальным взлетом цен на газ, акции Яндекса вышли на докризисный уровень, что объяснимо снижением спроса на "бесконтактные" услуги.
